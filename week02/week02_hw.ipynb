Write Python code to do the following web scraping exercise with BeautifulSoup and lxml:
1. Connect to https://pythonprogramming.net/parsememcparseface/ using urllib.request
2. Create an object called 'soup' using the source
3. Print page title, get attributes, values, and beginning navigation with specific values
4. Use str(paragraph.text) to get all the paragraphs
5. Get all the links using url.get
6. Get all the text using get_text

```python
from bs4 import BeautifulSoup
import urllib.request

# 1. Connect to the URL
url = "https://pythonprogramming.net/parsememcparseface/"
req = urllib.request.Request(url)
html_content = urllib.request.urlopen(req).read()

# 2. Create BeautifulSoup object
soup = BeautifulSoup(html_content, 'lxml')

# 3. Print page title
print("Page Title:", soup.title.text)
print('\n')

# Get attributes and values of the beginning navigation
begin_nav = soup.nav
if begin_nav:
    print("Beginning Navigation Attributes and Values:")
    for attribute in begin_nav.attrs:
        print(f"{attribute}: {begin_nav[attribute]}")
else:
    print("No beginning navigation found.")
print('\n')

# 4. Get all paragraphs
paragraphs = soup.find_all('p')
for i, paragraph in enumerate(paragraphs):
    print(f"Paragraph {i+1}: {str(paragraph.text)}")
print('\n')

# 5. Get all links
links = soup.find_all('a')
print("All Links:")
for link in links:
    print(link.get('href'))
print('\n')

# 6. Get all text
all_text = soup.get_text()
print("\nAll Text:")
print(all_text)
```
